```python
liverpool_team_id=40
API_BASE="https://v3.football.api-sports.io"
API_HEADERS={"x-apisports-key": API_KEY, "Accept": "application/json"}
```

```python
def fetch_data(team_id: int) -> Dict[str, Any]:
    url = f"{API_BASE}/transfers"
    resp = requests.get(url, params={"team": team_id}, headers=API_HEADERS, timeout=30)
    resp.raise_for_status()
    return resp.json()
```

```python
from datetime import datetime, UTC

raw_json=fetch_data(liverpool_team_id)
timestamp = datetime.now(UTC).strftime("%Y%m%dT%H%M%SZ")

raw_json_path = dir_raw / f"transfers_team{liverpool_team_id}_{timestamp}.json"
with open(raw_json_path, "w", encoding="utf-8") as f:
    json.dump(raw_json, f, ensure_ascii=False, indent=2)
```

```python
def transform_json_to_csv(raw: Dict[str, Any]) -> pd.DataFrame:
    rows:List[Dict[str, Any]]=[]
    for item in raw.get("response", []):
        player=item.get("player", {}) or {}
        player_id = player.get("id")
        player_name = player.get("name")
        api_update = item.get("update")

        for tr in item.get("transfers", []) or []:
            date_str = tr.get("date")
            type_raw = tr.get("type")
            teams = tr.get("teams") or {}
            t_in = teams.get("in") or {}
            t_out = teams.get("out") or {}
            rows.append({
                "player_id": player_id,
                "player_name": player_name,
                "api_record_updated_at": api_update,
                "transfer_date_raw": date_str,
                "transfer_type_raw": type_raw,
                "destination_team_id": t_in.get("id"),
                "destination_team_name": t_in.get("name"),
                "destination_team_logo": t_in.get("logo"),
                "source_team_id": t_out.get("id"),
                "source_team_name": t_out.get("name"),
                "source_team_logo": t_out.get("logo"),
            })
    return pd.DataFrame(rows)
```

```python
df_raw=transform_json_to_csv(raw_json)
csv_raw_path = dir_raw / f"transfers_team{liverpool_team_id}_{timestamp}.csv"
df_raw.to_csv(csv_raw_path, index=False)

print("Provera da li je sacuvano u tabelu:")
df_raw.head(8)
```

```python
def parse_transfer_fee_and_type(type_text: Optional[str]) -> Tuple[str, Optional[float]]:
    if type_text is None:
        return ("Unknown", np.nan)
    s = str(type_text).strip().lower()
    map = {"loan": ("Loan", 0.0),
        "free": ("Free", 0.0),
        "n/a": ("Unknown", 0.0),
        "loan return": ("Loan", 0.0),
        "?": ("Unknown", 0.0),
        "undisclosed": ("Unknown", 0.0),
        "-": ("Unknown", 0.0),
        "—": ("Unknown", 0.0),
    }
    if s in map:
        return map[s]
    cleaned=s.replace(",", "").replace("€", "").replace("$", "").replace("£", "").strip()
    m = re.search(r"([0-9]*\.?[0-9]+)\s*([mk])?", cleaned)
    if m:
        val = float(m.group(1))
        suf = m.group(2) or ""
        mult = 1.0
        if suf == "m": mult = 1_000_000.0
        elif suf == "k": mult = 1_000.0
        return ("Transfer", val * mult)
    return ("Unknown", np.nan)
```

```python
df= df_raw.copy()
df["transfer_date"]=pd.to_datetime(df["transfer_date_raw"], errors="coerce")
df["transfer_season_year"]=df["transfer_date"].dt.year

parsed=df["transfer_type_raw"].apply(parse_transfer_fee_and_type)
df["transfer_kind"]=parsed.apply(lambda t: t[0])
df["transfer_fee_eur"]=parsed.apply(lambda t: t[1]).astype("float64")
```

```python
def edge_direction(row: pd.Series, liv_id: int=40)-> Tuple[Optional[int], Oprional[int]]:
    src = row.get("source_team_id")
    dst = row.get("destination_team_id")
    if src == liv_id and pd.notna(dst):
        return (liv_id, int(dst))
    if dst == liv_id and pd.notna(src):
        return (int(src), liv_id)
    return (None, None)
```

```python
direction=df.apply(edge_direction, axis=1, result_type="expand")
df["source_club_id"], df["target_club_id"]=direction[0], direction[1]

df_edges = df.dropna(subset=["source_club_id", "target_club_id"]).copy()
df_edges["source_club_id"] = df_edges["source_club_id"].astype(int)
df_edges["target_club_id"] = df_edges["target_club_id"].astype(int)
```

```python
# Da bi analiza bila konkretnija ogranicio sam sezone koje se posmatraju
# Pocetna sezona bice sezona 2004 a krajnja 2024

years=df_edges["transfer_season_year"].dropna().astype(int).sort_values().unique()
print("Trenutno sve godine", years.tolist())
```

```python
# Razlog uklanjanja 1999-2004 jer sadrze nekompletne podatke i 2025 - jer godine nemaju kompletne informacije o transferima u tim  sezonama
start_year=2004
end_year=2024
choosen_years= [y for y in years if start_year<=y<=end_year]
print("Godine u analizi:", ", ".join(str(int(y)) for y in choosen_years))
```

```python
edges_in_range = df_edges[df_edges["transfer_season_year"].isin(choosen_years)].copy()
print("Broj edgesa: ", len(edges_in_range))
```

```python
edges_in_range.head(5)
```

```python
id_to_name:Dict[int,str]={}

for _,r in edges_in_range.iterrows():
    source_id=int(r["source_club_id"]); source_name = r.get("source_team_name")
    target_id=int(r["target_club_id"]); target_name = r.get("destination_team_name")
    if pd.notna(source_name) and source_id not in id_to_name:
        id_to_name[source_id] = str(source_name).strip()
    if pd.notna(target_name) and target_id not in id_to_name:
        id_to_name[target_id] = str(target_name).strip()

id_to_name[liverpool_team_id]="Liverpool"
```

```python
all_ids=pd.unique(
    pd.concat([edges_in_range["source_club_id"], edges_in_range["target_club_id"]], ignore_index=True)
).astype(int)
```

```python
# odredjivanje sastava nodes.csv i edges.csv fajla
nodes_df = pd.DataFrame({
    "club_id": all_ids,
    "club_name": [id_to_name.get(int(tid), f"Club_{int(tid)}") for tid in all_ids]
}).sort_values("club_id")

edges_df = edges_in_range[[
    "player_id", "player_name",
    "transfer_date", "transfer_season_year",
    "transfer_kind", "transfer_fee_eur",
    "source_club_id", "target_club_id"
]].copy()

nodes_path = dir_processed / "nodes.csv"
edges_path = dir_processed / "edges.csv"
nodes_df.to_csv(nodes_path, index=False)
edges_df.to_csv(edges_path, index=False)
```

```python
#ponderi u sledecim sekcijama

edges_in_range["is_transfer"]=(edges_in_range["transfer_kind"] == "Transfer").astype(int)
edges_in_range["is_loan"]=(edges_in_range["transfer_kind"] == "Loan").astype(int)
edges_in_range["is_free"]=(edges_in_range["transfer_kind"] == "Free").astype(int)
```

```python
edges_aggregated=(
    edges_in_range
    .groupby(["source_club_id", "target_club_id"], as_index=False)
    .agg(
        transfers_count=("player_id", "count"),
        total_fee_eur=("transfer_fee_eur", "sum"),
        count_transfer=("is_transfer", "sum"),
        count_loan=("is_loan", "sum"),
        count_free=("is_free", "sum"),
        avg_fee_eur=("transfer_fee_eur", "mean")
    )
)

edges_aggregated["avg_fee_eur"] = edges_aggregated["avg_fee_eur"].fillna(0.0)
```

```python
aggregated_path=dir_processed / "edges_aggregated.csv"
edges_aggregated.to_csv(aggregated_path, index=False)
edges_aggregated.sort_values("transfers_count", ascending=False).head(10)
```

```python
graph = nx.DiGraph()                                    # usmeren graf

# Dodaj cvorove sa imenom (labela)
for _, row in nodes_df.iterrows():
    graph.add_node(int(row["club_id"]), club_name=row["club_name"])
    
main_weight = "transfers_count"

# Dodaj ivice i sve atribute
for _, row in edges_aggregated.iterrows():
    graph.add_edge(
        int(row["source_club_id"]),
        int(row["target_club_id"]),
        weight=float(row[main_weight]),                
        transfers_count=int(row["transfers_count"]),
        total_fee_eur=float(row["total_fee_eur"]),
        count_transfer=int(row["count_transfer"]),
        count_loan=int(row["count_loan"]),
        count_free=int(row["count_free"]),
        avg_fee_eur=float(row["avg_fee_eur"])
    )
```

```python
plt.figure(figsize=(16, 11)) 
pos = nx.spring_layout(graph, seed=42, k=0.5) 
nx.draw_networkx_nodes(graph, pos, node_size=500, node_color="lightblue") 
weights = [d["weight"] for _, _, d in graph.edges(data=True)] 
nx.draw_networkx_edges(graph, pos, width=[w*0.05 for w in weights], arrows=True, alpha=0.6) 

labels = nx.get_node_attributes(graph, "club_name") 
nx.draw_networkx_labels(graph, pos, labels, font_size=6) 
plt.title(f"Transfer graf (weight = {main_weight})") 
plt.axis("off") 
graph_ju_path = dir_fig / "oriented-graph-jupiter.png"
plt.savefig(graph_ju_path, dpi=160)
plt.show()
```

```python
# nakon kreiranja grafa prikazujem neke deskriptivne analize, klubo koji najvise posluju ka ili od liverpula
# jedno od ogranicenja ovde je sto imamo transfere izmedju klubova koji su unknown tipa, ali ti podaci nisu iskljuceni jer i dalje predstavljaju 
# bitan faktor analize trgovanja izmedju klubova
# 1. kome liverpul najvise prodaje/ pozajmljuje igrace
top_out = (
    edges_aggregated[edges_aggregated["source_club_id"] == liverpool_team_id]
    .sort_values("transfers_count", ascending=False)
    .head(10)
    .merge(nodes_df, left_on="target_club_id", right_on="club_id")
    [["target_club_id","club_name","transfers_count","total_fee_eur","count_loan","count_free","count_transfer"]]
)
print("TOP 10 klubova u koje igraci idu iz Liverpula")
display(top_out)
```

```python
#2. od koga liverpul najvise kupuje/pozajmljuje igrace
top_in = (
    edges_aggregated[edges_aggregated["target_club_id"] == liverpool_team_id]
    .sort_values("transfers_count", ascending=False)
    .head(10)
    .merge(nodes_df, left_on="source_club_id", right_on="club_id")
    [["source_club_id","club_name","transfers_count","total_fee_eur","count_loan","count_free","count_transfer"]]
)
print("TOP 10 klubova od kojih Liverpul najvise kupuje/pozajmljuje igrace:")
display(top_in)
```

```python
#Metrike centralnosti

#in and out degree
in_deg_w=dict(graph.in_degree(weight="weight"))
out_deg_w=dict(graph.out_degree(weight="weight"))

#betweenness
betw=nx.betweenness_centrality(graph, weight="weight", normalized=True)

#page rank
pr=nx.pagerank(graph,weight="weight")

#closeness
harm_closeness=nx.harmonic_centrality(graph)

centralities_df = pd.DataFrame({"club_id": list(graph.nodes())})
centralities_df["in_degree_weighted"]  = centralities_df["club_id"].map(in_deg_w)
centralities_df["out_degree_weighted"] = centralities_df["club_id"].map(out_deg_w)
centralities_df["betweenness"]         = centralities_df["club_id"].map(betw)
centralities_df["pagerank"]            = centralities_df["club_id"].map(pr)
centralities_df["harmonic_closeness"]  = centralities_df["club_id"].map(harm_closeness)
centralities_df = centralities_df.merge(nodes_df, on="club_id", how="left")

centralities_path = dir_processed / "centralities.csv"
centralities_df.to_csv(centralities_path, index=False)
```

```python
# zajednice - ovde ocekujem da ce svi biti deo jedne zajednice jer je ipak to zvezda sema skoro, zato su svi okupljnei oko liverpula
import community as community_louvain 
```

```python
undirected_graph=graph.to_undirected()
partition=community_louvain.best_partition(undirected_graph, weight="weight", resolution=1.0)

nx.set_node_attributes(graph, partition, "community_id")

communities_df=(
    pd.DataFrame({"club_id": list(partition.keys()), "community_id": list(partition.values())})
    .merge(nodes_df, on="club_id", how="left")
)

communities_df["is_liverpool"] = communities_df["club_id"].eq(liverpool_team_id)
```

```python
## Trendovi po sezonama- analiziracemo neke trendove ponasanja liverpula tokom sezona
temp=edges_in_range[[
    "transfer_season_year", "transfer_fee_eur", "source_club_id", "target_club_id"
]].copy()

temp["to_liverpool"] = (temp["target_club_id"] == liverpool_team_id).astype(int)
temp["from_liverpool"] = (temp["source_club_id"] == liverpool_team_id).astype(int)

agg_transfers=(
    temp.groupby("transfer_season_year")
       .agg(transfers_in=("to_liverpool", "sum"),
            transfers_out=("from_liverpool", "sum"))
       .reset_index()
)

temp["fee_in"] = np.where(temp["to_liverpool"] == 1, temp["transfer_fee_eur"].clip(lower=0), 0.0)
temp["fee_out"] = np.where(temp["from_liverpool"] == 1, temp["transfer_fee_eur"].clip(lower=0), 0.0)

agg_fee = (
    temp.groupby("transfer_season_year")
       .agg(fee_in_sum=("fee_in", "sum"),
            fee_out_sum=("fee_out", "sum"))
       .reset_index()
)
```

```python
trend = pd.merge(agg_transfers, agg_fee, on="transfer_season_year", how="outer").fillna(0)
trend = trend[trend["transfer_season_year"].isin(choosen_years)].sort_values("transfer_season_year")
```

```python
trend["transfers_statistics"] = (trend["transfers_in"] / (trend["transfers_out"] + 1e-9)).round(3)
trend["fee_statistics"]  = (trend["fee_in_sum"] / (trend["fee_out_sum"] + 1e-9)).round(3)

trend_path = dir_processed / "trend_liverpool.csv"
trend.to_csv(trend_path, index=False)
```
